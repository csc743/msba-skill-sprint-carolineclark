```python

todo_list = [
    {"task": "Ship package", "done": False},
    {"task": "Go to yoga class", "done": False},
    {"task": "Pick up medicine", "done": False},
    {"task": "Go to doctor's appointment", "done": False},
    {"task": "Work", "done": False},
    {"task": "Go to class", "done": False}
]


def show_tasks():
    print("\nCAROLINES TO DO LIST <3")
    for i, item in enumerate(todo_list, start=1):
        status = "YES, GOOD JOB" if item["done"] else "DO IT!!"
        print(f"{i}. {item['task']} [{status}]")
    print()


def add_task():
    task = input("Enter your new task")
    todo_list.append({"task": task, "done": False})
    print("Task added!")


def complete_task():
    show_tasks()
    try:
        num = int(input("Which task did you do? "))
        todo_list[num - 1]["done"] = True
        print("WELL DONE")
    except:
        print("error")


def delete_task():
    show_tasks()
    try:
        num = int(input("Which task do you want to delete? "))
        removed = todo_list.pop(num - 1)
        print(f"Removed: {removed['task']}")
    except:
        print("error")

while True:
    print("\n--- HELLO CAROLINE ---")
    print("1. View tasks")
    print("2. Add task")
    print("3. Complete task")
    print("4. Delete task")
    print("5. Exit")

    choice = input("Choose an option: ")

    if choice == "1":
        show_tasks()
    elif choice == "2":
        add_task()
    elif choice == "3":
        complete_task()
    elif choice == "4":
        delete_task()
    elif choice == "5":
        print("Bye Bye")
        break
    else:
        print("error")
```
